```python
from datasets import Dataset, Audio
import os

wav_dir = "dataset/nptel-pure/wav"
txt_dir = "dataset/nptel-pure/corrected_txt"

data = []
for filename in os.listdir(wav_dir):
    if filename.endswith(".wav"):
        audio_path = os.path.join(wav_dir, filename)
        transcript_path = os.path.join(txt_dir, filename.replace(".wav", ".txt"))
        if os.path.exists(transcript_path):
            with open(transcript_path, "r", encoding="utf-8") as f:
                transcription = f.read().strip()
            data.append({"audio": audio_path, "text": transcription})

dataset = Dataset.from_list(data)
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))
```

```python
from datasets import DatasetDict
import random

# Shuffle and split (90% train, 10% test)
random.shuffle(data)
split_idx = int(0.9 * len(data))
train_data = data[:split_idx]
test_data = data[split_idx:]

# Rebuild datasets from splits
train_dataset = Dataset.from_list(train_data).cast_column("audio", Audio(sampling_rate=16000))
test_dataset = Dataset.from_list(test_data).cast_column("audio", Audio(sampling_rate=16000))

# Combine into a DatasetDict
final_dataset = DatasetDict({
    "train": train_dataset,
    "test": test_dataset
})
```
